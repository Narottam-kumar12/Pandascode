```python
import pandas as pd
df = pd.DataFrame()
```

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
tempDf = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
tempDf.to_csv('batsman_runs.csv',index = False)
```

```python
df.pivot_table(index = 'batsman',columns='bowling_team',values = 'batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

```python
tempDf.to_excel("batsman_runs.xlsx")
```

```python
tempDf.to_excel("output.xlsx",sheet_name = 'batsman_runs')
```

```python
tempDf2 = df.pivot_table(index = 'batsman',columns='bowling_team',values = 'batsman_runs',aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    tempDf.to_excel(writer,sheet_name = 'Sheet_name_1')
    tempDf2.to_excel(writer,sheet_name = 'Sheet_name_2')
```

```python
df.query('batsman_runs == 6').pivot_table(index = 'over',columns = 'ball',values = 'batsman_runs',aggfunc = 'count').to_html('sixes_heatmap.html')
```

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack().to_json('ipl.json')
```
